```python
import pandas as pd
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt

df = pd.read_csv("EV.csv")
df.head()
```

```python
df = pd.DataFrame(df)
print(type(df))
print(df.columns)
print(df.head())
```

```python
df['number_of_cells'].fillna('0', inplace=True)
df['towing_capacity_kg'].fillna(df['towing_capacity_kg'].mean(), inplace=True)
df['torque_nm'].fillna(df['torque_nm'].mean(), inplace=True)
df.dropna(subset=['fast_charging_power_kw_dc'], inplace=True)
df.dropna(subset=['cargo_volume_l'], inplace=True)
df.dropna(subset=['model'], inplace=True)
```

```python
df.to_csv("cleaned_data.csv", index= False)
cleaned_data = pd.read_csv("cleaned_data.csv")
print(cleaned_data.head())
```

```python
print(cleaned_data.isnull().sum())
print("\n\n", cleaned_data.duplicated().any())
```

```python
cleaned_data = pd.read_csv("cleaned_data.csv")
print(cleaned_data.head())
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
print(numeric_cols)
```

```python
from sklearn.preprocessing import Normalizer, MinMaxScaler, StandardScaler, RobustScaler

normalizer = Normalizer()
normalized_data = pd.DataFrame(normalizer.fit_transform(cleaned_data[numeric_cols]), columns=numeric_cols)
print("\nNormalized Data (first 5 rows):\n", normalized_data.head())


# Min-Max Scaler
minmax_scaled = pd.DataFrame(MinMaxScaler().fit_transform(cleaned_data[numeric_cols]), columns=numeric_cols)
print("\nMin-Max Scaled Data (first 5 rows):\n", minmax_scaled.head())

# Standard Scaler
standard_scaled = pd.DataFrame(StandardScaler().fit_transform(cleaned_data[numeric_cols]), columns=numeric_cols)
print("\nStandard Scaled Data (first 5 rows):\n", standard_scaled.head())

# Robust Scaler
robust_scaled = pd.DataFrame(RobustScaler().fit_transform(cleaned_data[numeric_cols]), columns=numeric_cols)
print("\nRobust Scaled Data (first 5 rows):\n", robust_scaled.head())

# Central Tendency and Dispersion
central_dispersion = {
    'Mean': cleaned_data[numeric_cols].mean(),
    'Median': cleaned_data[numeric_cols].median(),
    'Mode': cleaned_data[numeric_cols].mode().iloc[0],
    'Std Dev': cleaned_data[numeric_cols].std(),
    'Variance': cleaned_data[numeric_cols].var(),
    'Range': cleaned_data[numeric_cols].max() - cleaned_data[numeric_cols].min(),
    'IQR': cleaned_data[numeric_cols].quantile(0.75) - cleaned_data[numeric_cols].quantile(0.25)
}
central_dispersion_df = pd.DataFrame(central_dispersion)
print("\nCentral Tendency and Dispersion:\n", central_dispersion_df)
```
